```python
import sys
import pprint
import xlsxwriter
import json
from dataclasses import dataclass
from typing import ClassVar


@dataclass
class FileManagement:
    json_file_name: str = "/Users/khadidjabaaziz/Desktop/dashoard/dash_01.json"
    excel_file_name: str = "output_dash_01.xlsx"
    page: ClassVar[list] = [] #dictionnaire 
    widgets: ClassVar[list] = []
    widgetRequest : ClassVar[list] = []
        
    def read_text_file(self):
        try:
            with open(self.json_file_name) as data_file:
                data = json.load(data_file)
                page = data["page"]
                title = page["title"]
                widgets = data["widgets"]
                widgetRequest = data["widgetRequest"]
                #pprint.pprint(title)
                for dic in widgets:
                    w_id = dic["_id"]#1
                    w_etag = dic["_etag"]#1
                    canBeCompared = dic["canBeCompared"]#1
                    chartType = dic["chartType"]#1
                    conditions = dic["conditions"]#vide
                    widgetType = dic["widgetType"]#1
                    pageId = dic["pageId"]#1
                    title = dic["title"]#1
                    widgetCategory = dic["widgetCategory"]#1
                    request = dic["request"]#plus
                    for select in request["selectAdvanced"]:
                        for k, v in select.items():
                            return k, '/',v
                    data = dic["data"]#plusi
                    groupBy = dic["groupBy"]#plus
                    for elt in groupBy[0]:
                        return elt, '/',groupBy[0][elt]
                    category = dic["category"]#2
                    for k, v in category.items():
                        print( k,v)
                    widgetProps = dic["widgetProps"] #plusi
                    for k, v in widgetProps.items():
                        return k, '/', v 
                        for el in k['colorSetTheme']:
                            return el
                    
                
        except:
            print("Unexpected error : ", sys.exc_info()[0])
            raise
    
            
if __name__ == '__main__':
    file_management = FileManagement()
    file_management.read_text_file()
    #file_management.save_to_xlsx()
 
```
